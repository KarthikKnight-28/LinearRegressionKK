```python
from sklearn.model_selection import train_test_split
```

```python
feature=d[['median_house_value']]
```

```python
label=d.drop(d[['median_house_value']],axis=1)
```

```python
label
```

```python
feature
```

```python
x_train,x_test,y_train,y_test=train_test_split(label,feature,test_size=5,random_state=12)
```

```python
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

```python
from sklearn import linear_model
```

```python
model=linear_model.LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
model.score(x_test,pred)
```
